```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
```

```python
plt.figure()
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth = 10) #라인 굵기 = linewidth
```

```python
plt.plot(x, y, color = 'red') #컬러 = color
```

```python
plt.plot(x, y, marker = 'o', markersize = 10) #점 표시 : marker, 점 크기 : markersize
```

```python
plt.plot(x, y, linestyle = ':')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'bo') #b : blue, o : 점
```

```python
plt.plot(x, y)
plt.title('Markerplay')
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x, y)
plt.grid(True)
```
